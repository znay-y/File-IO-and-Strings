## Question 1 [0.75 marks] – Reading from a File

Write a program that:
- Reads a text file called `input.txt`
- Prints each line to the console

You must use `BufferedReader`.


```java
// Write your code here
import java.io.BufferedReader;
import java.io.FileReader;
import java.io.IOException;

  public static void main(String[] args) {
    try (BufferedReader reader = new BufferedReader(new FileReader("text.txt"))) {
      String line;
      while ((line = reader.readLine()) != null) {
        System.out.println(line);
      }
    } catch (IOException e) {
      System.out.println("Error reading file.");
    }
  }
```

## Question 3 [0.75 marks] – Basic Regex Matching

Write a method:

`static boolean isValidEmail(String email)`

that returns `true` if the input is a valid email and `false` otherwise.

Use a regular expression.


```java
// Write your code here
static boolean isValidEmail(String email){
        String regex = "^[A-Za-z0-9+_.-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,}$";
        if (email.matches(regex)) {
            return true;
        } else {
            return false;
        }
    }
```

## Question 4 [0.75 marks] – Extracting Data with Regex

Write a method:

`static ArrayList<String> extractNumbers(String text)`

that extracts all numbers from a string using regex and returns them as a list.


```java
// Write your code here
static ArrayList<String> extractNumbers(String text) {
        ArrayList<String> Aaray = new ArrayList<String>();
        String regex = "^[0-9]";
        String[] letters = text.split("");

        for (int i = 0; i < letters.length; i++) {
            if (letters[i].matches(regex)) {
                Aaray.add(letters[i]);
            }
        }

        return Aaray;
    }
```
